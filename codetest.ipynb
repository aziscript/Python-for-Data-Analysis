```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Read in the new salaries dataset
salaries = pd.read_csv('salaries.csv')

salaries.head()
```

```python
# Get the month of the response
salaries["month"] = pd.to_datetime(salaries["date_of_response"]).dt.month

# Extract the weekday of the response
salaries["weekday"] = pd.to_datetime(salaries["date_of_response"]).dt.weekday
```

```python
for col in salaries.columns:
    if salaries[col].dtype == 'object':
        salaries[col] = salaries[col].astype('category').cat.codes
```

```python
# Create a heatmap
sns.heatmap(salaries.corr(), annot=True)
plt.show()
```
